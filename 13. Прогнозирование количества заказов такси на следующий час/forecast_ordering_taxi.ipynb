Обернем график в функцию

```python
def get_plot(dataframe, feature_names, title, rolling_mean_window=0, rolling_std_window=0):
    """
    Функция строит график для любого количества признаков
    dataframe: датафрейм
    feature_names: список признаков
    title: название графика
    rolling_mean_window: размер окна для скользящего среднего (по умолчанию 0 - скользящее среднее не строится)
    rolling_std_window: размер окна для скользящего стандартного отклонения (по умолчанию 0 - скользящее стандартное отклонение не строится)
    """
    fig, ax = plt.subplots(figsize=(15, 5))
    for feature in feature_names:
        ax.plot(dataframe.index, dataframe[feature], label=feature)
    if rolling_mean_window > 0:
        for feature in feature_names:
            rolling_mean = dataframe[feature].rolling(window=rolling_mean_window).mean()
            ax.plot(dataframe.index, rolling_mean, label=f'{feature} rolling mean ({rolling_mean_window})')
    if rolling_std_window > 0:
        for feature in feature_names:
            rolling_std = dataframe[feature].rolling(window=rolling_std_window).std()
            ax.plot(dataframe.index, rolling_std, label=f'{feature} rolling std ({rolling_std_window})')
    ax.legend()
    ax.set_title(title)
    ax.set_xlabel('Даты')
    ax.set_ylabel('Количество заказов')
    ax.minorticks_on()
    ax.grid(which='major', linewidth=.5)
    ax.grid(which='minor', linewidth=.25, linestyle='--')
    plt.show()
```

### Декомпозиция

Напишем функцию для декомпозиции:

```python
def decomposition_plot(*args):
    """
     Функция принимает:
     *args: временные ряды.
     Возвращает:
     Графическое представление декомпозиции временных рядов в виде набора подграфиков, объединенных в одном окне.

     Для каждого временного ряда создается отдельный подграфик на одном окне. На каждом графике отображается временной ряд,
     соответствующий компоненту декомпозиции (тренд, сезонность или остаток), а также заданы название графика и легенда для
     временного ряда. На графиках также заданы оси координат, сетка и отображаются мажорные и минорные деления на осях.

     """
    fig, axes = plt.subplots(nrows=3, figsize=(15, 10))
    fig.suptitle('Декомпозиция для количества заказов', fontsize=16)

    for n, p in enumerate(args):
        ax = axes[n]
        ax.plot(p, lw=1, label=p.name)
        ax.set_title(p.name)
        ax.legend()
        ax.minorticks_on()
        ax.grid(which='major', linewidth=.5)
        ax.grid(which='minor', linewidth=.25, linestyle='--')
    plt.tight_layout()
    plt.show()
```

Посмотрим декомпозицию за август:

### Cтатистический тест Дики-Фуллера (Dickey-Fuller test)

Для проверки стационарности временного ряда используем статистический тест Дики-Фуллера (Dickey-Fuller test): Это один из наиболее распространенных тестов на стационарность. Он используется для проверки гипотезы о наличии единичного корня во временном ряду. Если результат теста показывает значимость, то ряд можно считать нестационарным. Наиболее распространенные значения уровня значимости включают 0.05 (5%) и 0.01 (1%). Это означает, что если p-значение, полученное в результате теста Дики-Фуллера, меньше выбранного уровня значимости, то гипотеза о наличии единичного корня будет отвергнута, и временной ряд можно считать стационарным.

Формулируем гипотезы:

$H_0$: Временной ряд имеет единичный корень, то есть он нестационарен.

$H_1$: Временной ряд не имеет единичного корня, то есть он стационарен.

```python
def adf_test(col, alpha=0.05):
    """
    Функция проверяет статистическую гипотезу о стационарности временного ряда при помощи ADF-теста.
    Выводит результаты теста на стационарность.
    col: Временной ряд (список, массив или Series)
    alpha: Уровень значимости (по умолчанию 0.05)
    """
    result = adfuller(col)
    p_value = result[1]
    
    print(f'p-value = {p_value:.4f}')
    
    if p_value <= alpha:
        print('Отвергаем нулевую гипотезу о наличии единичного корня. Временной ряд стационарен.')
    else:
        print('Не удалось отвергнуть нулевую гипотезу о наличии единичного корня. Временной ряд нестационарен.')
```

Зададим уровень значимости 0.05

```python
adf_test(data['num_orders'], alpha=0.05)
```

Зададим уровень значимости 0.01

```python
adf_test(data['num_orders'], alpha=0.01)
```

Если временной ряд имеет слабые стационарные свойства, то его свойства могут меняться в зависимости от уровня значимости. Например, если ряд имеет небольшой тренд, то при более высоком уровне значимости мы можем не отвергнуть нулевую гипотезу о наличии единичного корня, тогда как при более низком уровне значимости мы можем ее отвергнуть. Таким образом, выбор уровня значимости может стать причиной изменения результата теста. Однако из графика выше мы все-же можем предположить, что наш временной ряд нестационарен.

### Автокорреляция

Построим график автокорреляции временного ряда для обнаруженного сезона в 24 лага и запакуем в функцию:

```python
def autocorr_plot(col, lags, figsize=None, color='b', lw=1, grid=True):
    """
    Функция строит график автокорреляции временного ряда.
    col: pandas.Series - временной ряд
    lags: int - количество лагов для расчета автокорреляции
    figsize: tuple - размер графика (по умолчанию None - адаптивный размер)
    color: str - цвет линий автокорреляции (по умолчанию 'b')
    lw: int - толщина линий автокорреляции (по умолчанию 1)
    grid: bool - показывать сетку на графике или нет (по умолчанию True)
    """
    fig, ax = plt.subplots(figsize=figsize)
    plot_acf(col, lags=lags, marker='o', linestyle='--', lw=lw, ax=ax, color=color)
    ax.set_xlabel('Лаг')
    ax.set_ylabel('Значение автокорреляции')
    ax.set_title(f'Автокорреляция {col.name}')
    if grid:
        ax.minorticks_on()
        ax.grid(which='major', linewidth=.5)
        ax.grid(which='minor', linewidth=.25, linestyle='--')
    plt.show()
```

Построим график суточной автокорреляции

Обнаруженные значения автокорреляции > 0.5 на 1-м, 23-м и 24-м лаге подтверждают наличие зависимости в данных. Это означает, что значения заказов такси в конкретный час связаны с предыдущими значениями заказов в течение последних суток. Если в предыдущие дни в определенное время было много заказов, то скорее всего такая же тенденция будет и в текущий день в то же самое время. Это подтверждает наличие сезонности в имеющихся у нас данных.

### Создадим новые признаки на основе даты и времени:

Напишем функцию:

```python
def create_features(dataframe, max_lag, rolling_size):
    """
    Функция создает новые признаки на основе данных входного датафрейма.

    Аргументы:
    dataframe -- исходный датафрейм с данными о количестве заказов
    max_lag -- максимальное значение сдвига для создания признаков-лагов
    rolling_size -- размер окна для расчета скользящих средних и стандартных отклонений

    Возвращает:
    df_copy -- датафрейм с добавленными признаками
    """

    # Создаем копию исходного датафрейма, чтобы не изменять его
    df_copy = dataframe.copy()

    # Добавляем признаки, связанные с временем
    # Час
    df_copy['hour'] = df_copy.index.hour  
    # День
    df_copy['day'] = df_copy.index.day  
    # День недели
    day_of_week_bins = pd.cut(df_copy.index.dayofweek, bins=[-1, 0, 1, 2, 3, 4, 5, 6], labels=[
        'monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday'], include_lowest=True)
    df_copy['day_of_week'] = day_of_week_bins.astype(str)  
    # Время суток
    time_bins = pd.cut(df_copy.index.hour, bins=[0, 6, 12, 18, 24], labels=[
                       'night', 'morning', 'afternoon', 'evening'], include_lowest=True)
    df_copy['time_of_day'] = time_bins.astype(str)
    # Бинарный признак для выходных дней (суббота и воскресенье)
    df_copy['weekend'] = df_copy.index.dayofweek.isin([5, 6]).astype(int)  

    # Добавляем признаки, связанные с количеством заказов
    # Количество заказов за предыдущий день
    df_copy['orders_prev_day'] = df_copy['num_orders'].shift(24)  
    # Скользящее среднее по размеру окна rolling_size
    df_copy[f'rolling_mean_{rolling_size}'] = df_copy[
        'num_orders'].shift(1).rolling(rolling_size).mean()  
    # Скользящее среднее по размеру окна rolling_size*7
    df_copy[f'rolling_mean_{rolling_size*7}'] = df_copy[
        'num_orders'].shift(1).rolling(rolling_size*7).mean()  
    # Стандартное отклонение по размеру окна rolling_size
    df_copy[f'rolling_std_{rolling_size}'] = df_copy[
        'num_orders'].shift(1).rolling(rolling_size).std()  
    # Стандартное отклонение по размеру окна rolling_size*7
    df_copy[f'rolling_std_{rolling_size*7}'] = df_copy[
        'num_orders'].shift(1).rolling(rolling_size*7).std()  

    # Добавляем признаки-лаги
    for lag in range(1, max_lag + 1):
        df_copy[f'lag_{lag}'] = df_copy['num_orders'].shift(lag)

    return df_copy
```

### Подготовка к обучению и тестированию  моделей

Разобьем датафрейм на выборки и удалим пропуски из тренировочной выборки. Сделаем тестовую выборку размером 10% от исходных данных.

```python
train, test = train_test_split(data, shuffle=False, random_state=RANDOM_STATE, test_size=0.1)
train = train.dropna()

X_train = train.drop(['num_orders'], axis=1)
X_test = test.drop(['num_orders'], axis=1)

y_train = train['num_orders']
y_test = test['num_orders']
```

Проверим разбиение:

```python
print(f'Количество строк и столбцов в обучающей выборке: {X_train.shape}, '
 f'что составляет {round(len(X_train)/len(data)*100, 2)} % от данных.')
print(f'Количество строк и столбцов в тестовой выборке: {X_test.shape}, '
 f'что составляет {round(len(X_test)/len(data)*100, 2)} % от данных.')
```

Объявим списки с числовыми и категориальными признаками:

```python
cat_features = list(X_train.select_dtypes(include=['category', 'object']).columns)
num_features = list(X_train.select_dtypes(include=['number']).columns)
```

Создаем объекты с OneHotEncoder, ColumnTransformer, MinMaxScaler

```python
preprocessor_cat = OneHotEncoder(handle_unknown='ignore')

preprocessor_linear = ColumnTransformer(
    transformers=[('cat', preprocessor_cat, cat_features),
                  ('num', Pipeline([('poly', PolynomialFeatures(degree=2)),
                                    ('scaler', MinMaxScaler())]), num_features)])
preprocessor_boosting = ColumnTransformer(
    transformers=[('cat', preprocessor_cat, cat_features),
                  ('num', PolynomialFeatures(degree=2), num_features)])
```

Создадим таблицу с результатами обучения моделей:

```python
result_table = pd.DataFrame(columns=['Название модели', 'Лучшие гиперпараметры', 'RMSE'])
```

Объявим функцию, которая будет тестировать модели и возвращать результаты и экземпляр Pipeline с лучшими параметрами

```python
def test_model(pipeline, param_grid):
    """
    Функция test_model выполняет оценку и тестирование модели машинного обучения с использованием метода 
    кросс-валидации и поиска оптимальных гиперпараметров. Она принимает два аргумента:
    pipeline: объект pipeline, содержащий последовательность шагов обработки данных и модель для обучения. 
    Модель должна быть доступна через ключ 'regressor' в named_steps атрибуте pipeline.
    param_grid: словарь с набором гиперпараметров для перебора во время поиска оптимальных значений.
    """
    # Получаем название модели из pipeline
    model_name = pipeline.named_steps['regressor'].__class__.__name__
    
    # Создаем объект GridSearchCV для текущей модели
    grid_search = GridSearchCV(pipeline, param_grid=param_grid, cv=TimeSeriesSplit(n_splits=5),
                               scoring='neg_root_mean_squared_error',
                               refit=True, error_score='raise', verbose=10, n_jobs=-1)

    # Обучаем модель на тренировочных данных
    grid_search.fit(X_train, y_train)

    # Получаем лучшие параметры и создаем новый экземпляр Pipeline с лучшими параметрами
    best_params = grid_search.best_params_
    pipeline.set_params(**best_params)

    # Обучаем модель с лучшими параметрами
    best_estimator = grid_search.best_estimator_

    # Предсказываем
    prediction = np.round(np.abs(grid_search.predict(X_train)))
    
    # Получаем метрику RMSE
    RMSE = abs(grid_search.best_score_)

    # Сохраняем результаты кросс-валидации для текущей модели
    result = {'Название модели': model_name,
          'Лучшие гиперпараметры': best_params,
          'RMSE': round(RMSE, 2)}

    return result, best_estimator, best_params, prediction
```

### LinearRegression

```python
%%time

# Создаем объект модели
model = LinearRegression()

# Определяем параметры сетки для подбора гиперпараметров
param_grid = {}

# Создаем Pipeline с моделью
pipeline = Pipeline([
    ('preprocessor', preprocessor_linear),
    ('regressor', model)
])

# Вызываем функцию test_model
lr_result, lr_estimator, lr_best_params, lr_pred = test_model(
    pipeline, param_grid)

# Выводим результаты
display(lr_result)

# Добавляем результаты в сводную таблицу
result_table = result_table.append(lr_result, ignore_index=True)
```
