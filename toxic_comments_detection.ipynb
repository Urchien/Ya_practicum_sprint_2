Интернет-магазин «Викишоп» запускает новый сервис. Теперь пользователи могут редактировать и дополнять описания товаров, как в вики-сообществах. То есть клиенты предлагают свои правки и комментируют изменения других. Магазину нужен инструмент, который будет искать токсичные комментарии и отправлять их на модерацию. 

**Задача:** Обучить модель классифицировать комментарии на позитивные и негативные со значением метрики качества F1 не меньше 0.75. В нашем распоряжении набор данных с разметкой о токсичности правок. 

**Описание данных:**

Данные находятся в файле [toxic_comments.csv](https://code.s3.yandex.net/datasets/toxic_comments.csv). 
Столбец text в нём содержит текст комментария, а toxic — целевой признак.

1. Подготовить данные для машинного обучения

```python
def get_train_test_split(dataframe, test_size=.25, test_embed_size=.90, random_state=RANDOM_STATE):
    """
    Функция принимает на вход исходный датафрейм и параметры для разбиения 
    на обучающую и тестовую выборки. Дополнительно она формирует из 10% 
    обучающей выборки выборку на эмбеддинги. 
    Функция выводит информацию о размерах выборок и доле единиц в таргетах 
    для обучения и тестирования, а также для выборки на эмбеддинги.

    Аргументы функции:
    dataframe: исходный датафрейм
    test_size: доля тестовой выборки (по умолчанию 0.25)
    test_embed_size: доля выборки на эмбеддинги от X_train (по умолчанию 0.90)
    random_state: random state для разбиения (по умолчанию RANDOM_STATE)

    Возвращаемые значения:
    X_train: обучающая выборка, содержащая признаки
    X_test: тестовая выборка, содержащая признаки
    y_train: обучающая выборка, содержащая целевую переменную
    y_test: тестовая выборка, содержащая целевую переменную
    X_embed: выборка для эмбеддингов, содержащая признаки
    y_embed: выборка для эмбеддингов, содержащая целевую переменную
    """
    # Разбиваем исходный датасет на train и test
    X_train, X_test, y_train, y_test = train_test_split(
        dataframe.drop('toxic', axis=1), dataframe['toxic'],
        stratify=dataframe['toxic'], test_size=test_size, random_state=random_state)

    # Формируем выборку на эмбеддинги из X_train
    X_embed, _, y_embed, _ = train_test_split(
        X_train, y_train, stratify=y_train, test_size=test_embed_size, random_state=random_state)

    # Выводим информацию о размерах выборок и доле единиц в train и test
    print(
        f'Количество строк и столбцов в обучающей выборке: {X_train.shape}, '
        f'что составляет {X_train.shape[0] / dataframe.shape[0]:.2%} от данных.\n'
        f'Доля единиц в обучающем таргете: {sum(y_train == 1) / sum(y_train == 0):.2%}\n'
        f'Количество строк и столбцов в тестовой выборке: {X_test.shape}, '
        f'что составляет {X_test.shape[0] / dataframe.shape[0]:.2%} от данных.\n'
        f'Доля единиц в тестовом таргете: {sum(y_test == 1) / sum(y_test == 0):.2%}\n'
        f'Количество строк и столбцов в обучающей выборке (эмбеддинги): {X_embed.shape}, '
        f'что составляет {X_embed.shape[0] / X_train.shape[0]:.2%} от обучающей выборки.\n'
        f'Доля единиц в обучающем таргете(эмбеддинги): {sum(y_embed == 1) / sum(y_embed == 0):.2%}'
    )

    return X_train, X_test, y_train, y_test, X_embed, y_embed
```

```python
def get_embeddings(feature, batch_size=256, seed=42):
    """
    Функция получает на вход Series текстовых признаков и возвращает их эмбеддинги, 
    полученные с помощью предобученной модели BERT. 

    Аргументы:
    feature: pandas Series со строками текста, для которых нужно получить эмбеддинги
    batch_size: размер пакета (batch size) для обработки текстов. По умолчанию 256.
    seed: начальное значение для генератора случайных чисел numpy. По умолчанию 42.

    Возвращает: numpy массив эмбеддингов размерности (n_samples)
    """

    # Установка seed для повторяемости результатов
    np.random.seed(seed)

    # Загрузка предобученного токенизатора и модели BERT
    tokenizer = transformers.BertTokenizer.from_pretrained('bert-base-uncased')
    model = transformers.BertModel.from_pretrained('bert-base-uncased')

    # Создание копии входных данных
    feature_emb = feature.copy()

    # Токенизация текстов
    tokenized = feature_emb.apply(
        lambda x: tokenizer.encode(x, add_special_tokens=True, truncation=True, max_length=512))

    # Определение максимальной длины среди всех закодированных текстов
    max_len = 0
    for i in tokenized.values:
        if len(i) > max_len:
            max_len = len(i)

    # Добавление паддинга до максимальной длины
    padded = np.array([i + [0]*(max_len - len(i)) for i in tokenized.values])

    # Создание attention mask
    attention_mask = np.where(padded != 0, 1, 0)

    # Инициализация списка для хранения эмбеддингов
    embeddings = []

    # Вычисление общего количества пакетов
    total_batches = padded.shape[0] // batch_size

    # Цикл по пакетам для создания эмбеддингов
    for i in notebook.tqdm(range(total_batches)):
        # Создание пакета из токенизированных признаков
        batch = torch.LongTensor(padded[batch_size*i:batch_size*(i+1)])
        # Создание маски внимания для текущего пакета
        attention_mask_batch = torch.LongTensor(attention_mask[batch_size*i:batch_size*(i+1)])
        # Создание тензора из нулей с размерностью текущего пакета для token_type_ids
        token_type_ids_batch = torch.zeros_like(batch)
        # Отключение градиентного вычисления внутри блока кода, чтобы уменьшить использование памяти
        with torch.no_grad():
        # Вычисление эмбеддингов для текущего пакета при помощи загруженной модели BERT
        # В данном случае используется только выходное значение последнего слоя (параметр last_hidden_state).
        # Это позволяет получить векторное представление (эмбеддинг) каждого токена в последовательности признаков.
        # Размерность полученного тензора равна (batch_size, max_len, hidden_size), 
        # где hidden_size - размер скрытого слоя модели BERT.
        # При этом, если max_len в текущем пакете меньше максимальной длины признаков, то для оставшихся элементов тензора
            batch_embeddings = model(input_ids=batch,
                                     attention_mask=attention_mask_batch,
                                     token_type_ids=token_type_ids_batch)
        
        # Цикл по пакетам для создания эмбеддингов
        embeddings.append(batch_embeddings[0][:, 0, :].numpy())
        # Освобождение памяти, занятой тензорами
        del batch, attention_mask_batch, token_type_ids_batch
        gc.collect()

    # Обработка последнего пакета, если его размер меньше batch_size
    if padded.shape[0] % batch_size != 0:
        remaining_samples = padded.shape[0] - (total_batches * batch_size)
        remaining_batch = torch.LongTensor(padded[total_batches * batch_size:])
        remaining_attention_mask = torch.LongTensor(attention_mask[total_batches * batch_size:])
        remaining_token_type_ids = torch.zeros_like(remaining_batch)
        with torch.no_grad():
            remaining_embeddings = model(input_ids=remaining_batch, 
                                         attention_mask=remaining_attention_mask,
                                         token_type_ids=remaining_token_type_ids)
        
        embeddings.append(remaining_embeddings[0][:remaining_samples, 0, :].numpy())
        # Освобождение памяти, занятой тензорами
        del remaining_batch, remaining_attention_mask, remaining_token_type_ids
        gc.collect()

    # Объединение всех полученных эмбеддингов в один массив
    features_ = np.concatenate(embeddings)
    # Возвращение эмбеддингов
    return features_
```

```python
def save_load_file(variable, file_name, folder_name='project_data', download=False):
    """
    Функция сохраняет или загружает файлы локально в формате pickle.
    
    Аргументы:
    variable: переменная, которую нужно сохранить или загрузить
    file_name: имя файла
    folder_name: имя папки, в которую будут сохранены файлы. По умолчанию 'project_data'.
    download: флаг, указывающий, нужно ли загружать файл. По умолчанию False - сохранение файла.
    
    Если download=True, то функция загружает файл и возвращает переменную.
    Если download=False, то функция сохраняет переменную в файл.
    """
    
    os.makedirs(folder_name, exist_ok=True)
    
    # Если download=True, то загрузить файл и вернуть переменную
    if download:
        with open(os.path.join(folder_name, f'{file_name}.pickle'), 'rb') as f:
            return pickle.load(f)
    
    # Если download=False, то сохранить переменную в файл
    else:
        with open(os.path.join(folder_name, f'{file_name}.pickle'), 'wb') as f:
            pickle.dump(variable, f)
```

```python
def download_from_link(url: str, file_path: str):
    """Загружает данные из pickle-файлов по доступной ссылке для скачивания в переменные"""
    
    # Отправляем GET-запрос по указанному URL
    response = requests.get(url)

    # Открываем файл в бинарном режиме для записи
    with open(file_path, 'wb') as file:
        # Записываем полученные данные в файл
        file.write(response.content)

    # Открываем файл в бинарном режиме для чтения
    with open(file_path, 'rb') as file:
        # Загружаем данные из файла
        data = pickle.load(file)

    # Выводим сообщение об успешной загрузке файла
    print(f'Done! Файл {file_path} загружен в переменную!')
    return data
```

```python
def preprocess_and_tokenize(text, cache=None):
    """
    Принимает исходный текст и возвращает его очищенную и лемматизированную версию.
    Аргументы:
        text (str): Исходный текст для обработки.
        cache (dict, optional): Кэш для хранения лемм слов. По умолчанию None.
    Возвращает:
        str: Очищенный и лемматизированный текст.
    """
    # Проверка на пустую строку или неправильный тип аргумента
    if not isinstance(text, str) or len(text) == 0:
        return ""
    # Создаем экземпляр лемматизатора WordNet
    lemmatizer = WordNetLemmatizer()  
    # Загружаем стоп-слова на английском языке
    stop_words = set(stopwords.words('english'))  
    # Преобразуем текст к нижнему регистру и удаляем знаки пунктуации и цифры
    text = text.lower().translate(str.maketrans('', '', string.punctuation + string.digits))  
    # Токенизация текста
    tokens = word_tokenize(text)  

    tag_map = {
        'N': wordnet.NOUN,
        'V': wordnet.VERB,
        'R': wordnet.ADV,
        'J': wordnet.ADJ
    }

    words_lem = []  # Список для хранения лемм слов
    added_words = set()  # Множество для отслеживания уже добавленных лемм

    for token in tokens:
        if len(token) > 1 and token not in stop_words:  # Проверка длины токена и наличия в стоп-словах
            if cache is None or token not in cache:  # Проверяем, есть ли лемма токена в кэше
                pos_tag = nltk.pos_tag([token])[0][1][0].upper()  # Определение части речи токена
                pos = tag_map.get(pos_tag, wordnet.NOUN)  # Конвертация тега части речи в формат WordNet
                lem_token = lemmatizer.lemmatize(token, pos)  # Лемматизация токена

                if cache is not None:
                    cache[token] = lem_token  # Добавляем лемму токена в кэш

            else:
                lem_token = cache[token]  # Если лемма уже есть в кэше, извлекаем ее

            if lem_token not in added_words:  # Проверяем, не была ли лемма уже добавлена
                words_lem.append(lem_token)  # Добавляем лемму в список
                added_words.add(lem_token)  # Добавляем лемму в множество уже добавленных слов

    preprocessed_text = " ".join(words_lem)  # Преобразуем список лемм в строку, разделенную пробелами
    return preprocessed_text  # Возвращаем очищенный и лемматизированный текст
```

```python
def get_scatter(X, y, title, svd=True):
    """
    Функция для визуализации данных на плоскости с использованием метода SVD.
    Аргументы:
    X - матрица признаков размерности (n_samples, n_features), где n_samples - количество объектов, а n_features - количество признаков.
    y - вектор меток классов размерности (n_samples).
    title - заголовок графика.
    svd - флаг, указывающий, нужно ли использовать метод SVD для снижения размерности данных (по умолчанию True).
    Возвращаемое значение:
    График с точками на плоскости, где каждая точка представляет объект из набора данных, 
    цвет точки соответствует метке класса, а размер точки зависит от значения вектора y 
    (если значение метки класса равно 1, то размер точки равен 2, в противном случае равен 1).
    """
    if svd:
        # Снижение размерности данных до 2 компонент с помощью метода SVD
        svd = TruncatedSVD(n_components=2)
        X = svd.fit_transform(X)

    # Задание размеров точек на графике в зависимости от метки класса
    sizes = [2 if label == 1 else 1 for label in y]

    # Создание объектов фигуры и осей графика с заданным размером
    fig, ax = plt.subplots(figsize=(15, 8))
    # Рисование точек на графике
    scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.get_cmap('RdYlBu_r'), s=sizes)
    # Задание заголовка графика, а также скрытие делений на осях
    ax.set_title(title)
    ax.set_xticks([])
    ax.set_yticks([])

    # Создание списка объектов типа Line2D, представляющих элементы легенды на графике
    legend_elements = [
        plt.Line2D([0], [0], marker='o', color='w', markersize=8,
                   markerfacecolor=scatter.get_cmap()(1.0), label='Токсичные комментарии'),
        plt.Line2D([0], [0], marker='o', color='w', markersize=8,
                   markerfacecolor=scatter.get_cmap()(0.0), label='Нормальные комментарии')
    ]
    # Добавление легенды на график
    ax.legend(handles=legend_elements)
    # Отображение графика
    plt.show()
```

```python
def test_model(X_train, y_train, model, param_dist, random_state=RANDOM_STATE, embed=True):
    """
    Функция test_model выполняет оценку и тестирование модели машинного обучения с использованием метода 
    кросс-валидации и поиска оптимальных гиперпараметров. 
    Аргументы:
    X_train: обучающая выборка, содержащая признаки
    y_train: обучающая выборка, содержащая целевую переменную
    model: алгоритм модели машинного обучения
    random_state: random state для разбиения (по умолчанию RANDOM_STATE)
    embed: флаг для вида данных выборки
    param_dist: словарь с набором гиперпараметров для перебора во время поиска оптимальных значений.
    """
    # Создаем Pipeline с моделью
    pipeline = Pipeline([('classifier', model)])
    # Создаем объект RandomizedSearchCV для подбора гиперпараметров
    search = RandomizedSearchCV(pipeline, param_distributions=param_dist,
                                n_iter=12, scoring='f1', n_jobs=-1, cv=3, 
                                verbose=10, random_state=random_state)
    # Обучаем объект RandomizedSearchCV на обучающей выборке
    search.fit(X_train, y_train)

    # Извлекаем лучшую модель и ее гиперпараметры
    best_model = search.best_estimator_
    best_params = search.best_params_
    model_name = search.best_estimator_.named_steps['classifier'].__class__.__name__
    # Вычисляем метрики на кросс-валидации
    prediction = search.predict(X_train)
    f1_scores = search.best_score_
    precision_scores = precision_score(y_train, prediction)
    recall_scores = recall_score(y_train, prediction)
    # Выводим вид данных по флагу
    features = 'embedding' if embed else 'tf_idf'
    # Выводим название модели, лучшие гиперпараметры и метрики

    result = {'Данные': features, 
              'Mодель': model_name,
              'Гиперпараметры': best_params,
              'Precision': round(precision_scores, 2),
              'Recall': round(recall_scores, 2),
              'F1 на train': round(f1_scores, 2)}
    # Возвращаем лучшую модель и метрики
    return result, best_model
```

```python
def plot_confusion_matrix_and_roc_curve(y_true, y_pred_prob):
    # Вычисляем матрицу ошибок
    y_pred = y_pred_prob > 0.5
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    # Создаем ROC-кривую
    fpr, tpr, thresholds = roc_curve(y_true, y_pred_prob)
    roc_auc = roc_auc_score(y_true, y_pred_prob)
    # Строим график
    fig, ax = plt.subplots(1, 2, figsize=(14, 7))
    # График матрицы ошибок
    sns.heatmap([[tn, fp], [fn, tp]], cmap="Blues", annot=True, fmt="d", cbar=False, ax=ax[0])
    ax[0].set_xticklabels([0, 1])
    ax[0].set_yticklabels([1, 0])
    ax[0].set_xlabel('Предсказанные значения')
    ax[0].set_ylabel('Истинные значения')
    ax[0].set_title(f'Матрица ошибок\n[TN, FP]\n[FN, TP]')
    # График ROC-кривой
    ax[1].plot(fpr, tpr, label='ROC-кривая (S = {:.2f})'.format(roc_auc))
    ax[1].plot([0, 1], [0, 1], 'k--')
    ax[1].set_xlabel('FPR')
    ax[1].set_ylabel('TPR')
    ax[1].set_title('ROC-кривая')
    ax[1].legend(loc="lower right")
    plt.show()
```
